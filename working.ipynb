## 1. Постановка задачи
Первая формулировка задачи:
- Понять, что я должен получить от заказчика, чтобы решить задачу, установить реалистичные сроки.

Уточнение задачи:
- Что заказчик ожидает увидеть в выводах исследования?
- Что нужно посчитать и зачем?
- Какие факторы (временные, внешние, политические, социальные и т.д) нужно учесть, чтобы они не искажали результаты?
- Каким способом анализировать полученные данные?
- Где нужно остановиться?

## 2. Получение данных
- Сформировать запрос на выгрузку данных к разработчикам, либо подготовить выгрузку самостоятельно.
- Убедиться, что нужны именно эти данные, и он представлены в форме, пригодной для анализа.

## 3. Подготовка данных
- Проверить данные на шумы, пропуски, выбросы, полноту и степень правдоподобия

## 4. Исследовательский анализ данных (EDA)
- Углублённое изучение данных.
- Визуализация данных.

## 5. Подготовка прототипа решения
- В прототипе используются только реальный контент, соответствующий теме продукта.
- Есть понимание того, что «сообщает» прототип без комментариев

## 6. Проверка результатов
- Оценить степень соответствия модели и её качество

## 7. Доработка решения
- Итеративный процесс анализа - повторять расчёт и валидацию результата до тех пор, пока не получится решение, качество которого устроит

## 8. Финальное решение и оформление результата
- Оформить финальное решение с описанием результата.
- Обязательно объяснить, **почему** и **как** они помогают ответить на исходный вопрос 

## 9. Проверка реальностью
- Изучить практические результаты решения, принятого на основе анализа данных.
- Сработало ли решение? Оказалось ли оно качественным и оправданным?
