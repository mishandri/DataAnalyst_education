Одномерные графики и анализ рассматривают одну переменную за раз и независимо от других. Мы обычно в основном интересуемся распределением ее значений и игнорируем другие признаки в нашем наборе данных.

В прошлом модуле мы уже строили **линейный график**, чтобы посмотреть отток пользователей во времени. <br>
Это были числа или по-другому **количественные данные**, а именно **дискретные**. Количественные данные обычно используются для  подсчета или измерение какой-либо переменной. <br>

**Дискретная величина** же это такая величина, которая может принимать значения только из _некоторого списка определенных чисел_. В большинстве случаев, так как список чисел определен и события происходят с определенной вероятностью, то наша величина _счетна и конечна_. Примеры в жизни: броски кубиков, число детей в семье, число звонков в службу поддержки, число клиентов с картой лояльности, число товаров в чеке. Заметим, что _дискретная величина это целое число_ и не принимает никакого интервального значения, что ее отличает от следующего типа количественных данных.<br>

Второй репрезентацией количественных данных является **непрерывная величина** (иногда могут называть интервальными данными). Это такая величина, которая в результате эксперимента принимает все значения из некоторого числового промежутка и _число возможных значений непрерывной случайной величины бесконечно_.<br>
Самый просто пример это рост: от 140.5 до 205.5 см и средний рост будет равен допустим 173.<br>
Если мы применим эти правила к оттоку клиентов, то говорить что мае ушло в отток 35.5 человек будет нерелевантно и из разряда **"полтора землекопа"**, тк количество людей обычно измеряются целыми числами. <br>
_Соответственно можно утверждать, что рост это непрерывная величина, а отток пользователей это дискретная величина (потому что количество людей, которые ушли в отток ограничено размером аудиториии пользователей нашего приложения/сервиса)._

Так же есть данные, которые регистрируют определенное качество, которым обладает объект или разграничивают объекты по категориям. Такие данные называют **качественными** или **категориальными**. Самыми простыми примерами _является цвет M&Ms, классы на круизе типа "Титаник" (1,2,3 классы), виды вина (красное, белое, розовое) и тд._<br>

В случае если у нас присутствуют только 2 значения (как с полом выше), такие данные называют **бинарными**. Даже если значениям этого качества можно прописать с помощью числа (например мужина-1 и женщина-0), то обрабатывать эти числа как количественные данные никак нельзя, тк с точки зрения машины/компьютера 1 обладает большим весом/смыслом, чем 0. <br>Бинарыне переменные очень широко используется при построение признаков в машинном обучении.<br>

Так же стоит отметить, что качественные/категориальные данные бывают двух типов: **порядковые**, для которых существует содержательный смысл при упорядовачивании, и **номинативные**, для которых нет содержательно интерпретируемого порядка. <br>
Это мы рассмотрим на примере ниже:)

Перейдем непосредственно к видам графиков.

Как мы обсуждали выше, отток это дискретная величина, принимающее какое-то целое значение и для такой величины подходит линейный график или ```lineplot```. Также тут можно добавить, что мы изображаем **изменение величины в течение какого-либо времени**.

Когда мы работаем с другими типами данных, например с **категориальнымими**, как в отчете о среднем чеке, где у нас было много категорий и видов категорий (MCC коды, категории покупок и видов валют), бывает полезными использовать **столбчатые диаграммы**, чтобы посмотреть количество наблюденй в каждой категории. В ```matplotlib``` они называются ```barplot```.<br> 

Посмотрим примеры на классическом датасете с отзывами о разных видах вин.<br>

````python
reviews = pd.read_csv('../data/winemag-data_first150k.csv.gz',compression='gzip' )
````

````python
reviews.head(3)
````

### Столбчатая диаграмма (Bar chart)

Тут мы посчитали количество наблюдений в нашей выборке, разбив их по категории ```province``` и топ-10. <br>
Также мы можем повернуть график в горизонтальную ориентацию используя  ```.bar()``` вместо  ```.barh()```.

````python
reviews['province'].value_counts().head(10).plot.bar()
#reviews['province'].value_counts().head(10).plot(kind='bar')
````

````python
reviews['province'].value_counts().head(10).plot.barh()
#reviews['province'].value_counts().head(10).plot(kind='barh')
````

Количество посчитано, но очень часто нужно банально **посчитать проценты от тотала по категории**.<br>
Это можно сделать, разделив наш топ-10 на длину всего датасета, вот так:

````python
(reviews['province'].value_counts().head(10) / len(reviews)*100).plot.bar()
````

**Столбчатые диагарамы** очень пластичный вид графиков, тк с помощью него можно представить любую информацию, где высота отображается по оси Y, а по оси X у нас категории.<br>

Помимо этого категории могут быть как "настоящими" или **номинальными категориями** (в нашем датасете это _страна происхождения вина, провинция, винодельня и тд._, в реальной жизни это могут быть различные _коды, например MCC или валют, почтовые индексы, имена людей и прочее_). Номинальные категории нет смысла упорядочивать, тк каждая категория имеет равное значение. <br>

**_Пример:_** Если мы строим график по количеству съеденых яблок для Маши, Даши и Пети, где количество яблок будет по оси Y, а имена детей по оси X, то в данном случае, нам нет смысла каким либо образом упорядовачивать имена детей. <br>

Интереснее становится, когда мы хотим проанализировать **порядковую категорию**, где есть какая-то внутренняя зависимость, например как в нашем случае с оценками для каждого вина ниже.<br>
Если мы посчитаем сколько вин набрало какое количество баллов, то метод ```value_counts()``` отсортирует результаты по значеним (от большего к меньшему по умолчанию), но при этом в индексах будет полная белиберда, тк после 88 балла идет сразу 90, а после него 86. Но мы же знаем что баллы имеют нарастающую характеристику- **чем выше оценка, тем выше и заведение**. Все что нам нужно сделать это отсортировать по индексу и построить ```barchart```<br>

````python
reviews['points'].value_counts().head()
````

````python
reviews['points'].value_counts().plot.bar()
````

````python
reviews['points'].value_counts().sort_index().head()
````

````python
reviews['points'].value_counts().sort_index().plot.bar()
````

### Линейный график (Line chart)

Хорошо, в нашем случае было проставлено всего 20 возможных оценок, но что если были бы все **100**? <br>
Каждый из столбцов в нашей диаграме был бы очень очень маленький и график был бы очень непонятным. <br>
Что в данном случае можно сделать? <br>
Можно, конечно, увеличить наш график, но это поможет до опредленного момента. В таком случае придется заменить столбчатую диаграмму на другой вид графика, а именно ```linechart``` и отрисовать его с помощью метода ```.plot.line()```<br>

````python
reviews['points'].value_counts().sort_index().plot.line(color='green')
````

Визуализация, используя линейный график это самый первый и базовый метод метод визуализации данных, так как подходит для широкого спектра случаев, в том числе, где столбчатые диаграммы становятся громоздкими, номинативными данными, а также для данных с временным фактором. <br>
Также подходит чтобы визуализировать разные срезы или категории и сопоставлять их между собой:

````python
reviews[reviews['province'] == 'Northern Spain']['points'].value_counts().sort_index().plot.line(color='green')
reviews[reviews['province'] == 'Northeastern Italy']['points'].value_counts().sort_index().plot.line(color='blue')
````

### График с областями (Area chart)

Также мы можем закрасить площадь под линейным графиком и унас уже получится **график с областями**. <br>
Бывает полезным при построение небольших графиков с несколькими областями, которые перекрывают друг друга.

````python
reviews['points'].value_counts().sort_index().plot.area()
````

### Гистограмма (Histogram)

Гистограмма или гистограмма частот вобщем-то является столбчатым графиком, а точнее ее **частным случаем**. <br>
Единственное отличие состоит в том, что **значения, которые мы передаем на вход в гистограмму сначала разбиваются** на <br>
N равных диапазонов/интервалов, а далее подсчитывается **сколько наблюдений попало в тот или иной диапазон**.

Здесь нам придется вспомнить еще один тип данных- **непрерывные величины**, тоесть те, которые не только имеют зависимость при порядке, но и еще могут быть измерены. Иногда их также называют интервальными, тк поддаются интервальной оценке, как раз посредством гистограммы.

````python
reviews['price'].plot.hist()
````

Но что-то не очень похоже на информативный график. Проблема заключается в том, что гистограмма очень восприимчива к экстремальным значениям, так как интервалы разбиения у нас строятся,учитывая разброс наших данных.<br>

Попробуем убрать выбросы. Для этого можно воспользоваться кодом из 3 урока, если мы хотим вычяислить определенный квантиль.<br>
Сейчас просто воспользуемся методом```.describe()``` и уберем все что выше 75ого квантиля, тоесть все значения, которые выше 75% от цены в нашем распределении.

````python
reviews.describe()
````

````python
reviews[reviews['price'] < 40]['price'].plot.hist()
````

Получили распрделение по ценам от минимального до 75-ого квартиля (тоесть от 4 до 40 долларов).<br>

Также можно построить гистограмму для порядковых значений, например как баллы. <br>
Так как в баллах у нас нет разброса, то нам не нужно убирать экстрималные значения/выбросы и 
мы можем сразу увидеть распределение, довольно похожее на нормальное с небольшим смещением в правую сторону.

````python
reviews['points'].plot.hist()
````

Итак, давайте подведем итоги того что мы узнали про построение одномерных графиков. Мы узнали что данные бывают **количественные** и **категориальные**. <br>

Количественные данные подразделяются на **непрерывные величины** и **дискретные величины**. <br>
- **Непрерывные величины** принимают значения из некоторого промежутка, число возможных значений непрерывной случайной величины бесконечно и поддается некоторой интервальной оценке.
- **Дискретная величина** же это  величина, множество значений которой конечно или счётно. Очевидно, значения дискретной случайной величины не содержат какой-либо непрерывный интервал на числовой прямой.<br>

Категориальные переменные в свою очередь можно подразделить на **номинальные** и **порядковые**. <br>
- **Номинальными категориями** (в нашем датасете можетбыть страна происхождения вина, провинция, винодельня тоесть в данном случе, категории нет смысла упорядочивать, тк каждая категория имеет равный вклад и значение/происходит с определенной вероятностью. Например, почтовый индекс, регион, религия. 
- **Порядковые категории** это такой тип данных, где нужно упорядочивать категории, так как есть какая-то внутренняя зависимость и одна имеет выше приоритет/вероятность, нежели чем другая. Например, баллы в шкале оценки работы сотрудника банка, баллы силы землетрясения или оценка в баллах за тест.

## Диаграмма рассеивания (Scatter plot)

Диаграмма рассеивания это самый просто способ репрезентации 2х переменных на 2D графике. Для того чтобы его составить нужно использовать метод ```plot.scatter()```  и передать 2 переменные, которые мы хотим соотнести на плоскости.

````python
reviews.plot.scatter(x='price', y='points')
````

Получилось довольно плотная диаграмма. Мы можем избавиться от выбросов в виде догого вина и посмотреть например только вина менее 100 долларов. 

````python
reviews[reviews['price'] < 100].plot.scatter(x='price', y='points')
````

Опять же, в нашем диапазоне слишком много наблюдений. Мы можем выбрать 100, 200, 500, 1000 наблюдений, чтобы график выглядел более разрежаным и более вменяемым.

````python
reviews[reviews['price'] < 100].sample(500).plot.scatter(x='price', y='points')
````

На графике можно выявить некую зависимость, что чем дороже вино тем больше балов оно получает, но сказать что между переменными выражена сильная корреляция не представляется возможность, тк есть вина с 92 баллами как за 100, так и за 20 с небольшим долларов.<br>

Заметьте, что нам пришлось снижать выборку и рисовать только ограниченное количество наблюдений. <br>
Это особенность данного типа графиков, которая показывает, что диаграммы рассеивания хорошо работают с выборками <br>
**небольшого размера** и где присутствует много **уникальных значений**. <br>

Если же у нас много неуникальных значений, то на место одной точки библиотека нарисует еще одну или несколько поверх и мы не увидим какую-то графическую репрезентацию большого количества наблюдений в одной точке,например более темный цвет или что-то подобное. <br>
Именно из-за этих моментов в таких условиях будет полезен следующий тип графиков.

## Шестигранная диаграмма/ Гексагон (Hex plot)

Шестигранная диаграмма метода ```.plot.hexbin()``` **агрегирует точки** в шесигранники и **использует цвет** для передачи интенсивности/количества наблюдений в той или иной области плоскости.<br>
Также на легенде графика рисуется некий "градусник", который показывает интенсивность палитры в зависимости от количества наблюдений.

````python
reviews[reviews['price'] < 100].plot.hexbin(x='price', y='points', gridsize=15)
````

Заметьте, что у нас нет оси X. Это лечится вот этим [трюком](https://stackoverflow.com/questions/43121584/matplotlib-scatterplot-x-axis-labels).

````python
fig,ax = plt.subplots()
reviews[reviews['price'] < 100].plot.hexbin(x='price', y='points', gridsize=15, ax=ax)
plt.xlabel('balls')
plt.show()
````

Этот график представляет все ту же информацию, что и диагармма рассеивания, но рассказывает уже другую историю.<br>
Теперь мы можем увидеть невооруженным взглядом, что большая часть наблюдений скопилась в области 87.5 пунктов и стоит около 20 долларов.<br>
Этого мы не смогли увидеть на графике с диаграммой рассеивания, тк наблюдений слишком много и точки перекрывали друг друга.

## Многослойная столбчатая диаграмма (Stacked bar chart)

Диаграммы рассеивания и шестигранные диаграмма это довольно экзотичные способы подачи информации и не все знакомы с такой формой представления. Но можно вернуться к более легким графикам и приспособить их под многомерность. Так например можно сделать с стобчатыми диаграммами и просто положить столбцы друг на друга. Такие графики довольно популярны среди сторонников экселя, тк они представляют основу визуализации во многих отчетах.

Давайте посмотрим как это работает в питоне.<br>
Я сделал небольшую предагрегацию, выделив топ-5 видов вина в нашем датасете япосчитал для каждого балла количество упоминаний по каждому из вин.

````python
# wine_counts = reviews[reviews['variety'].isin(list(reviews['variety'].value_counts().head(5).index))]\
#                     .groupby(['variety','points'])['points'].count().unstack().T
# wine_counts.to_csv("../data/wine_score_counts_5.csv")

wine_counts = pd.read_csv("../data/top-five-wine-score-counts.csv",index_col=0)
````

````python
wine_counts.head()
````

Мы можем построить многослойную столбчатую диаграмму просто передав методы `bar` аттрибут `stacked` со значением `True`, так как методы `matplotlib` ожидают данные именно в таком формате - одно разбиение по категорийной переменной в столбцам (виды вина) и одно разбиение на другую категорию (в нашем случае баллы или порядковую переменную) по строкам. <br>

````python
wine_counts.plot.bar(stacked=True)
````

Многослойные столбчатые диаграммы обладают все теми же плюсами и минусами, как и одномерные варианты - они неплохо справляются с категорийными признаками, например с номинативными и с небольшими объемами порядковых переменных. 

Другой график, который очень хорошо подходит под эти данные это много слойный график с областями. Здесь даже не нужно передавать
аттрибут `stacked` со значением `True`.

````python
wine_counts.plot.area()
````

Как и одномерные графики, многослойные предназначены для тех или иных типов данных - либо для категорийных, либо для количественных.<br>

Многослойные графики визуально более насыщенные и приятные глазу, однако обладают рядом проблем.
- Как мы уже говорили раньше, столбчатые диаграммы хорошо подходят для порядковых и номинативных данных небольшого объема. В данном случае 5 видов вин, которые являются номинативными переменными - золотая середина. Не слишком много, и не слишком мало- в самый раз для интерпретируемости графика. Ксожалению, нет какого-то определенного максимума для использования переменных в многослойных диаграммах, так как все крайне индивидуально, но на практике, заметно что более чем 8 признаков уже трудно отследить и соотнести друг с другом.<br>

- Другое ограничение вырастает из удобства многослойных графиков. На самом деле трудно понять конкретные цифры каждой категории, особенно если порядки категорий очень похожи. Например, сможете ли вы понять какое вино больше получило 87 баллов? <br>
Cabernet Sauvignon, Chardonnay или Pinot Noir?

## Многомерный линейный график (Bivariate line chart)

Один из, если не самый эффективный способ представления, все же, остается  линейный график. Из-за того что он занимает мало места он подходит и для многослойной визуализации.<br> 

````python
wine_counts.plot.line()
````

Используя линейные графики можно преодалеть ограничение столбчатых графиков в плане интерпретируемости, о котором мы говорили выше. 
Например, из графика выше мы видим, что самая популярна оценка среди топ-5 это 87 баллов, а так же что Chardonnay получило эту оценку больше всего раз.

Так же можно разбить отрисовку графиков по каждой категории отдельно, если мы не хотим нагромождать наш график.

````python
for i in wine_counts.columns:
    wine_counts[i].plot.line()
    plt.title(i)
    plt.show()
````
