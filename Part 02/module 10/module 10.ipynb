# Конструирование и выбор признаков

Нужно искать именно те признаки, которые описывают структуру объектов, которые мы рассматриваем.  
Признаки не должны дублировать друг друга.  
Хорошо спроектированный набор признаков даёт большую эффективность предсказаний.
Кроме того, необхлодимо выбирать оптимальные параметры и колиечство параметров для модели, иначе модель может срабатывать неправильно.  
Причём можно конструировать новые признаки на основе уже имеющихся.

Так же важно само представление данных, лучше всего подходят числовые данные, а не текстовые описания. Особенно, если они ранжируемы.
