### Задание 1.

Мы упоминали о том, что функция потерь AdaBoost имеет более жесткий штраф по отношению к ошибкам классификации, чем другие часто используемые функции. Реализуйте функции logloss и AdaBoost loss, которые возвращают значения штрафов, и посмотрите по графику, насколько сильно они отличаются (примем $x = y \hat{y}$).

- Logistic loss: $L(y, \hat{y}) = log(1 + exp(-y \hat{y}))$;
- AdaBoost loss: $L(y, \hat{y}) = exp(-y \hat{y})$.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

x = np.linspace(-2, 1, 300)

def logloss(x):
    return np.log(1 + np.exp(-x))

def AdaLoss(x):
    return np.exp(-x)

plt.figure(figsize=(10, 4))
plt.plot(x, logloss(x), label='logloss')
plt.plot(x, AdaLoss(x), label='AdaLoss')
plt.legend(loc='best')
plt.xlabel('X')
plt.ylabel('Loss')

x_check = np.array([-2, 0, 2])
y_log = np.array([2.127, 0.693, 0.127])
y_ada = np.array([7.389, 1., 0.135])

assert np.allclose(logloss(x_check), y_log, atol=1e-3), f'Function logloss is not verified'
assert np.allclose(AdaLoss(x_check), y_ada, atol=1e-3), f'Function AdaLoss is not verified'
```
